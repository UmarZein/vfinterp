```python
def gridspace(start, end, n, dim):
    linspace = torch.linspace(start, end, steps=n)
    mesh = torch.meshgrid(*[linspace] * dim, indexing='ij')
    grid = torch.stack(mesh, dim=-1).reshape(-1, dim)
    return grid
```

```python
def get_y(X):
    x=X[...,0]
    y=X[...,1]
    Y=torch.zeros_like(X)
    Y[...,0]=x.sin()-y.cos()
    Y[...,1]=y.sin()-x.cos()
    return Y
```

```python
class NoModel(MessagePassing):
    def __init__(self, k_dim, v_dim, hidden_size=120,
                 k=10,p=2):
        super().__init__(aggr='sum')
        self.k=k
        self.p=p
        
    def forward(self, K, V, Q, batch_kv=None, batch_q=None):
        if batch_kv is None:
            batch_kv=torch.zeros(K.shape[-2],device=K.device)
        if batch_q is None:
            batch_q=torch.zeros(Q.shape[-2],device=Q.device)
        src,tgt=edge_index = knn(K, Q, self.k, batch_x=batch_kv, batch_y=batch_q).flip(0)
        diffs = Q[tgt]-K[src]
        dists = diffs.norm(2,-1)+1e-8
        idists = dists.pow(-self.p)
        sumidists = scatter(idists, tgt, dim=-1, reduce="sum")[tgt]
        w = (idists/sumidists).view(-1,1)
        v=(V)
        out = self.propagate(
            edge_index, 
            v=(v,Q), 
            w=w,
            sizes=(K.shape[-2],Q.shape[-2])
        )#out: ?
        return out
        
    def message(self, v_j, w):
        return v_j*w
```

```python
class Decoder(MessagePassing):
    def __init__(self,k_dim,v_dim,hidden_size=120,k=10,p=2):
        super().__init__(aggr='sum')
        self.k=k
        self.p=p
        self.psi=nn.Sequential(
            nn.Linear(k_dim,hidden_size),
            nn.SiLU(),
            nn.Linear(hidden_size,v_dim),
            nn.SiLU(),
            nn.Linear(v_dim,v_dim),
        )
        self.phi=nn.Sequential(
            nn.Linear(v_dim,hidden_size),
            nn.SiLU(),
            nn.Linear(hidden_size,v_dim),
            nn.SiLU(),
            nn.Linear(v_dim,v_dim),
        )
        
    def forward(self, K, V, Q, batch_kv=None, batch_q=None):
        #K          f32:  N, k_dim
        #V          f32:  N, v_dim
        #Q          f32:  M, v_dim
        #batch_kv usize:  N
        #batch_q  usize:  M
        if batch_kv is None:
            batch_kv=torch.zeros(K.shape[-2],device=K.device)
        if batch_q is None:
            batch_q=torch.zeros(Q.shape[-2],device=Q.device)
        src,tgt=edge_index = knn(K, Q, self.k, batch_x=batch_kv, batch_y=batch_q).flip(0)
        diffs = Q[tgt]-K[src]
        phi_out = self.psi(diffs)
        dists = diffs.norm(2,-1)+1e-8
        idists = dists.pow(-self.p)
        sumidists = scatter(idists, tgt, dim=-1, reduce="sum")[tgt]
        w = phi_out*(idists/sumidists).view(-1,1)
        v=self.phi(V)
        out = self.propagate(
            edge_index, 
            v=(v,Q), 
            w=w,
            sizes=(K.shape[-2],Q.shape[-2])
        )
        return out
        
    def message(self, v_j, w):
        ret=v_j*w
        return ret
```

```python
class Decoder2(MessagePassing):
    def __init__(self,k_dim,v_dim,hidden_size=120,k=10,p=2):
        super().__init__(aggr='sum')
        self.k=k
        self.p=p
        self.phi=nn.Sequential(
            nn.Linear(k_dim+v_dim,hidden_size),
            nn.SiLU(),
            nn.Linear(hidden_size,v_dim),
            nn.SiLU(),
            nn.Linear(v_dim,v_dim),
        )
    def forward(self, K, V, Q, batch_kv=None, batch_q=None):
        #K          f32:  N, k_dim
        #V          f32:  N, v_dim
        #Q          f32:  M, v_dim
        #batch_kv usize:  N
        #batch_q  usize:  M
        if batch_kv is None:
            batch_kv=torch.zeros(K.shape[-2],device=K.device)
        if batch_q is None:
            batch_q=torch.zeros(Q.shape[-2],device=Q.device)
        src,tgt=edge_index = knn(K, Q, self.k, batch_x=batch_kv, batch_y=batch_q).flip(0)
        diffs = Q[tgt]-K[src]
        dists = diffs.norm(2,-1)+1e-8
        idists = dists.pow(-self.p)
        sumidists = scatter(idists, tgt, dim=-1, reduce="sum")[tgt]
        w = (idists/sumidists).view(-1,1)
        out = self.propagate(
            edge_index, 
            v=(V,Q), 
            w=w,
            diffs=diffs,
            sizes=(K.shape[-2],Q.shape[-2])
        )
        return out
        
    def message(self, v_j, w, diffs):
        ret= self.phi(torch.concat((v_j,diffs),-1))*w
        return ret
```

```python
import torch
import math

def map_features(diffs: torch.Tensor, 
                 B: torch.Tensor) -> torch.Tensor:
    """
    diffs: [N, D]
    B:    [F, D]
    
    returns: [N, 2*F] Fourier features of each diff vector
    """
    # 1) project to F frequencies
    #    (N×D) @ (D×F) → N×F
    proj = diffs @ B.t()      

    # 2) scale by 2π
    proj = 2 * math.pi * proj

    # 3) concat sin and cos → N×(2F)
    return torch.cat([proj.sin(), proj.cos()], dim=-1)
```

```python

class Decoder2FiLM(MessagePassing):
    def __init__(self, k_dim, v_dim, hidden_size=120, k=6, p=2, proj_out=None, fourier_B=None):
        super().__init__(aggr='sum')
        self.k = k
        self.p = p
        self.register_buffer("fourier_B",fourier_B)
        self.layer_norm = nn.LayerNorm(v_dim)
        
        # FiLM generator: map diffs (k_dim) → [γ(v_dim), β(v_dim)]
        self.film_gen = nn.Sequential(
            nn.Linear(2*self.fourier_B.shape[0], hidden_size),
            #nn.Linear(k_dim, hidden_size),
            nn.SiLU(),
            nn.Linear(hidden_size, 2 * v_dim)   # outputs γ||β
        )
        self.proj_out=proj_out

    def forward(self, K, V, Q, batch_kv=None, batch_q=None, diffs_scale=1.0, Y=None):
        if batch_kv is None:
            batch_kv = torch.zeros(K.size(0), device=K.device, dtype=torch.long)
        if batch_q  is None:
            batch_q  = torch.zeros(Q.size(0), device=Q.device, dtype=torch.long)

        # build k-NN graph
        src, tgt = edge_index = knn(K, Q, self.k,
                                   batch_x=batch_kv, batch_y=batch_q).flip(0)

        # relative positions & distances
        diffs = (Q[tgt] - K[src])*diffs_scale    # [#edges, k_dim]
        fourier_mapping = map_features(diffs, self.fourier_B)
        # print("decode diffs.std(-1).mean():",diffs.std(-1).mean())
        # print("decode diffs.mean():",diffs.mean())
        dists = diffs.norm(dim=-1) + 1e-8        # [#edges]
        idists = dists.pow(-self.p)
        sumid = scatter(idists, tgt, dim=0, reduce='sum')[tgt]
        w = (idists / sumid).view(-1, 1)         # [#edges, 1]

        out = self.propagate(
            edge_index,
            v=(V, Q),
            w=w,
            y=Y[edge_index[1]] if Y is not None else None,
            diffs=fourier_mapping,
            sizes=(K.size(0), Q.size(0)),
        )
        if Y is not None:
            return out/edge_index.shape[-1]
        return out

    def message(self, v_j, w, diffs, y=None):
        # Generate γ and β from diffs
        γβ = self.film_gen(diffs)               # [#edges, 2*v_dim]
        γ, β = γβ.chunk(2, dim=-1)              # each [#edges, v_dim]

        out = γ * v_j + β
        # apply FiLM: modulate v_j
        # (no need to unsqueeze if v_j is [#edges, v_dim])
        out = self.proj_out(self.layer_norm(out))

        # re-apply the attention weight
        if y is not None:
            return ((out - y)).pow(2)
        return out*w
```

```python

class Decoder3FiLM(MessagePassing):
    def __init__(self, k_dim, v_dim, hidden_size=120, k=6, p=2):
        super().__init__(aggr='mean')
        self.k = k
        self.p = p
        self.layer_norm = nn.LayerNorm(v_dim)
        
        # FiLM generator: map diffs (k_dim) → [γ(v_dim), β(v_dim)]
        self.film_gen = nn.Sequential(
            nn.Linear(k_dim, hidden_size),
            nn.SiLU(),
            nn.Linear(hidden_size, 2 * v_dim)   # outputs γ||β
        )

    def forward(self, K, V, Q, batch_kv=None, batch_q=None, diffs_scale=1.0):
        if batch_kv is None:
            batch_kv = torch.zeros(K.size(0), device=K.device, dtype=torch.long)
        if batch_q  is None:
            batch_q  = torch.zeros(Q.size(0), device=Q.device, dtype=torch.long)

        # build k-NN graph
        src, tgt = edge_index = knn(K, Q, self.k,
                                   batch_x=batch_kv, batch_y=batch_q).flip(0)

        # relative positions & distances
        diffs = (Q[tgt] - K[src])*diffs_scale    # [#edges, k_dim]
        # print("decode diffs.std(-1).mean():",diffs.std(-1).mean())
        # print("decode diffs.mean():",diffs.mean())

        return self.propagate(
            edge_index,
            v=(V, Q),
            diffs=diffs,
            sizes=(K.size(0), Q.size(0))
        )

    def message(self, v_j, diffs):
        # Generate γ and β from diffs
        γβ = self.film_gen(diffs)               # [#edges, 2*v_dim]
        γ, β = γβ.chunk(2, dim=-1)              # each [#edges, v_dim]

        out = γ * v_j + β
        # apply FiLM: modulate v_j
        # (no need to unsqueeze if v_j is [#edges, v_dim])
        out = self.layer_norm(out)

        # re-apply the attention weight
        return out
```

```python
class DownSampler(nn.Module):
    def __init__(self,
            ratio,
            mlp_dims_node: List[int],
            mlp_dims_edge: List[int],
            
            aggr: str = 'sum',
            
            node_norm=nn.LayerNorm, 
            final_node_norm=nn.Identity, 
            
            edge_norm=nn.LayerNorm,
            final_edge_norm=nn.Identity, 
            
            activation=nn.SiLU, 
            final_activation=nn.Identity, 
            
            dropout_rate=0.1, 
            final_dropout_rate=0.0,
            q=1,
            fourier_B=None,
            ):
        super().__init__()
        self.ratio=ratio
        self.q=q
        self.register_buffer("fourier_B",fourier_B)
        self.conv = ConvLayer(
            mlp_dims_node=mlp_dims_node,
            mlp_dims_edge=mlp_dims_edge,
            aggr=aggr,
            node_norm=node_norm,
            final_node_norm=final_node_norm,
            edge_norm=edge_norm,
            final_edge_norm=final_edge_norm,
            activation=activation,
            final_activation=final_activation,
            dropout_rate=dropout_rate,
            final_dropout_rate=final_dropout_rate,
        )
        self.layer_norm=nn.LayerNorm(mlp_dims_edge[-1])
        self.reset_parameters()
        
    def reset_parameters(self):
        self.conv.reset_parameters()
        
    def forward(self, K, V, batch=None, dK_scale=1.0, target_points=None, edge_index=None, sampled_idx=None):
        #K float:     N, k_dim
        #V float:     N, v_dim
        #batch int: N
        if batch is None:
            batch=torch.zeros(K.shape[-2],device=K.device,dtype=torch.long)
        if edge_index is None or sampled_idx is None:
            if sampled_idx is None:
                sampled_idx = fps(
                    K, 
                    ratio=self.ratio, 
                    batch=batch, 
                    random_start=True
                )
                edge_idx = knn(K[sampled_idx],K,1).flip(0)
                _, sampled_idx = scatter_min(torch.rand(edge_idx.shape[1],device=K.device), edge_idx[0], dim=0)
            edge_index = knn(
                K[sampled_idx], # src
                K, # tgt
                self.q, 
                batch_x=batch[sampled_idx], 
                batch_y=batch)
        src, tgt = edge_index
        dK = (K[sampled_idx][tgt]-K[src])*dK_scale
        fourier_mapping = map_features(dK, self.fourier_B)
        pooledV = self.conv(
            (V, torch.zeros_like(V[sampled_idx]) if target_points is None else target_points), 
            edge_index, 
            fourier_mapping)
        return K[sampled_idx], pooledV, batch[sampled_idx], edge_index, sampled_idx
```

```python
class UpSampler(nn.Module):
    def __init__(self,
            mlp_dims_node: List[int],
            mlp_dims_edge: List[int],
            
            aggr: str = 'sum',
            
            node_norm=nn.LayerNorm, 
            final_node_norm=nn.Identity, 
            
            edge_norm=nn.LayerNorm,
            final_edge_norm=nn.Identity, 
            
            activation=nn.SiLU, 
            final_activation=nn.Identity, 
            
            dropout_rate=0.1, 
            final_dropout_rate=0.0,
            fourier_B=None,
            ):
        super().__init__()
        self.conv = ConvLayer(
            mlp_dims_node=mlp_dims_node,
            mlp_dims_edge=mlp_dims_edge,
            aggr=aggr,
            node_norm=node_norm,
            final_node_norm=final_node_norm,
            edge_norm=edge_norm,
            final_edge_norm=final_edge_norm,
            activation=activation,
            final_activation=final_activation,
            dropout_rate=dropout_rate,
            final_dropout_rate=final_dropout_rate,
        )
        self.layer_norm=nn.LayerNorm(mlp_dims_edge[0])
        self.reset_parameters()
        self.register_buffer("fourier_B",fourier_B)
        
    def reset_parameters(self):
        self.conv.reset_parameters()
        
    def forward(self, K1, K2, V1, V2, edge_index, dK_scale=1.0):
        tgt, src = reversed_edge_index = edge_index.flip(0)
        dK = (K2[tgt]-K1[src])*dK_scale
        fourier_mapping = map_features(dK, self.fourier_B)
        informed_V1 = self.conv(
            (V2, V1), 
            reversed_edge_index, 
            fourier_mapping)
        return informed_V1
```

```python
class UNet(nn.Module):
    def __init__(self,
            ratios,
            mlp_dims_node: List[int],
            mlp_dims_edge: List[int],
            
            aggr: str = 'sum',
            
            node_norm=nn.LayerNorm, 
            final_node_norm=nn.Identity, 
            
            edge_norm=nn.LayerNorm,
            final_edge_norm=nn.Identity, 
            
            activation=nn.SiLU, 
            final_activation=nn.Identity, 
            
            dropout_rate=0.1, 
            final_dropout_rate=0.0,
            q=1,
            k_dim=2,
            fourier_B=None
            ):
        super().__init__()
        self.k_dim=k_dim
        self.register_buffer("fourier_B",fourier_B)
        self.downs = nn.ModuleList([])
        self.ups = nn.ModuleList([])
        self.n_layers = len(ratios)
        for ratio in ratios:
            down=DownSampler(
                ratio=ratio,
                mlp_dims_node=mlp_dims_node,
                mlp_dims_edge=mlp_dims_edge,
                aggr=aggr,
                node_norm=node_norm,
                final_node_norm=final_node_norm,
                edge_norm=edge_norm,
                final_edge_norm=final_edge_norm,
                activation=activation,
                final_activation=final_activation,
                dropout_rate=dropout_rate,
                final_dropout_rate=final_dropout_rate,
                q=q,
                fourier_B=fourier_B,
            )
            up=UpSampler(
                mlp_dims_node=mlp_dims_node,
                mlp_dims_edge=mlp_dims_edge,
                aggr=aggr,
                node_norm=node_norm,
                final_node_norm=final_node_norm,
                edge_norm=edge_norm,
                final_edge_norm=final_edge_norm,
                activation=activation,
                final_activation=final_activation,
                dropout_rate=dropout_rate,
                final_dropout_rate=final_dropout_rate,
                fourier_B=fourier_B,
            )
            self.downs.append(down)
            self.ups.append(up)
        self.reset_parameters()
        
    def reset_parameters(self):
        for d in self.downs:
            d.reset_parameters()
        for u in self.ups:
            u.reset_parameters()
        
    def forward(self, K, V, batch=None, history=None):
        res=[(K, V, batch, None, None)]
        if history is None:
            history = [(None, None, None, None, None) for _ in self.downs]
        dK_scale=torch.tensor(1.0)
        for i,(down, hist_entry) in enumerate(zip(self.downs, history)):
            hist_K, hist_V, hist_batch, hist_edge_index, hist_sampled_idx = hist_entry
            K, V, batch, edge_index, sampled_idx = out = down(
                K, V, batch, 
                dK_scale=dK_scale, 
                target_points=hist_V, 
                edge_index=hist_edge_index,
                sampled_idx=hist_sampled_idx
            )
            V=silu(V)
            history[i] = (K, V, batch, edge_index, sampled_idx)
            dK_scale=dK_scale*torch.tensor(down.ratio).pow(1/self.k_dim)
            if i+1<self.n_layers:
                res.append(out)
        for up, (_K, _V, _, _edge_index, _), down in zip(
            self.ups, 
            reversed(res), 
            reversed(self.downs)
        ):
            dK_scale=dK_scale*torch.tensor(down.ratio).pow(-1/self.k_dim)
            V=up(_K, K, _V, V, edge_index, dK_scale=dK_scale)
            V=silu(V)
            K=_K
            edge_index=_edge_index
        return V, history
```

```python
class VFInterpolator(nn.Module):
    def __init__(self,
            domain_dim,
            raw_feature_dim,
            latent_feature_dim,
            ratios,
            mlp_dims_node: List[int],
            mlp_dims_edge: List[int],
            
            aggr: str = 'sum',
            
            node_norm=nn.LayerNorm, 
            final_node_norm=nn.Identity, 
            
            edge_norm=nn.LayerNorm,
            final_edge_norm=nn.Identity, 
            
            activation=nn.SiLU, 
            final_activation=nn.Identity, 
            
            dropout_rate=0.1, 
            final_dropout_rate=0.0,

            decoder_hidden_size=120,
            k=10,
            q=1,
            p=2,
            n_encoders=1,
            key_scale=1.0,
            n_fouriers=100,
            fourier_std=100,
            ):
        super().__init__()
        assert mlp_dims_node[0]==mlp_dims_node[-1]==mlp_dims_edge[-1]==latent_feature_dim
        self.key_scale=key_scale
        self.proj_in =MLP(
            raw_feature_dim,
            [decoder_hidden_size],
            latent_feature_dim,
        )
        self.n_fouriers=n_fouriers
        fourier_B=torch.randn(n_fouriers,domain_dim)*fourier_std
        self.register_buffer("fourier_B",fourier_B)
        self.encoders=nn.ModuleList([])
        for _ in range(n_encoders):
            unet=UNet(
                    ratios=ratios,
                    mlp_dims_node=mlp_dims_node,
                    mlp_dims_edge=mlp_dims_edge,
                    aggr=aggr,
                    node_norm=node_norm,
                    final_node_norm=final_node_norm,
                    edge_norm=edge_norm,
                    final_edge_norm=final_edge_norm,
                    activation=activation,
                    final_activation=final_activation,
                    dropout_rate=dropout_rate,
                    final_dropout_rate=final_dropout_rate,
                    q=q,
                    k_dim=domain_dim,
                    fourier_B=self.fourier_B,
            )
            self.encoders.append(unet)
        #self.decoder=ExpressiveDecoder(domain_dim,latent_feature_dim,latent_feature_dim,hidden_size=decoder_hidden_size,k=k)
        #self.decoder=Decoder2(domain_dim,latent_feature_dim,hidden_size=decoder_hidden_size,k=k,p=p)
        self.decoder=Decoder2FiLM(
            domain_dim,
            latent_feature_dim,
            hidden_size=decoder_hidden_size,
            k=k,
            p=p,
            fourier_B=self.fourier_B,
            proj_out=MLP(
                latent_feature_dim,
                [decoder_hidden_size],
                raw_feature_dim,
                activation=Sin
            )
        )
        #self.decoder=Decoder(domain_dim,latent_feature_dim,hidden_size=decoder_hidden_size,k=k,p=p)
    def encode(self, K, V, batch_kv=None, batch_q=None):
        K=K*self.key_scale
        V = self.proj_in(V)
        history=None
        for encoder in self.encoders:
            V, history = encoder(K, V, batch_kv, history=history)
        return V
    def decode(self, K, V, Q, batch_kv=None, batch_q=None, Y=None):
        'when Y is not none, calculate the MSE instead'
        K=K*self.key_scale
        Q=Q*self.key_scale
        return (self.decoder(K, V, Q, batch_kv, batch_q, Y=Y))
    def forward(self, K, V, Q, batch_kv=None, batch_q=None, Y=None):
        V = self.encode(K, V, batch_kv=batch_kv, batch_q=batch_q)
        return self.decode(K, V, Q, batch_kv=batch_kv, batch_q=batch_q, Y=Y)
```

```python
import random
from torchvision.io import read_image
import os
from tqdm.auto import tqdm

all_train_files = []
for root, _, files in os.walk("./data/1pics/dude/"):
    for file in tqdm(files):
        all_train_files.append(os.path.join(root, file))
        
def parse_image(filename):
    #img=read_image("data/imagenette2-320/train/n01440764/ILSVRC2012_val_00002138.JPEG").permute(1,2,0)
    img=read_image(filename).permute(1,2,0)
    W, H, C = img.shape
    if C==1:
        C=3
        img = img.repeat(1,1,3)
        W, H, C = img.shape
    elif C!=3:
        raise Exception("bruh")
    # Flatten RGB values into [W*H, 3]
    Y = img.view(-1, C).float()  # [W*H, 3]
    
    # Generate pixel coordinate grid [W, H] -> [W*H, 2]
    x = torch.arange(W)
    y = torch.arange(H)
    xx, yy = torch.meshgrid(x, y, indexing='ij')  # 'ij' to match [W, H] layout
    
    X = torch.stack([xx, yy], dim=-1).float().view(-1, 2)  # [W*H, 2]
    
    X_=X.cuda()
    Y_=Y.cuda()/255
    return X_, Y_, W, H, C
    
import time
def parse_random_image():
    while True:
        try:
            return parse_image(random.choice(all_train_files))
        except:
            time.sleep(0.001)
```

```python
def calculate_density_kernel(pos, batch=None, k=15, sigma=1.0):
    # Get neighbors: edge_index shape [2, num_edges]
    edge_index = knn(pos, pos, k, batch_x=batch, batch_y=batch).flip(0)
    #edge_index, _ = remove_self_loops(edge_index)
    src, tgt = edge_index

    diffs = pos[src] - pos[tgt]
    dists_sq = (diffs ** 2).sum(dim=-1)  # squared Euclidean distance

    # Gaussian kernel weights (unnormalized)
    weights = torch.exp(-dists_sq / (2 * sigma ** 2)) + 1e-9  # add epsilon for stability

    # Normalize weights per source node so weights sum to 1 (mass-preserving)
    norm = scatter(weights, src, dim=0, dim_size=pos.size(0), reduce="sum")
    weights_norm = weights / norm[src]

    # Accumulate normalized weights on target nodes
    density = scatter(weights, tgt, dim=0, dim_size=pos.size(0), reduce="sum")

    return density
```

```python
def calculate_density(pos, batch=None, k=15, p=2,eps=1e-9):
    edge_index = knn(pos, pos, k, batch_x=batch, batch_y=batch).flip(0)#flip to get src-to-tgt
    edge_index, _ = remove_self_loops(edge_index)#remove self loops because inv(self_dist)=too big and noisy
    src,tgt=edge_index
    diffs = pos[src]-pos[tgt]
    dists = diffs.norm(2,-1)+eps
    idists = dists.pow(-p)
    density = scatter(idists, tgt, dim=-1, reduce="sum")#/k
    return density
```

```python
def calculate_density2(pos, batch=None, k=15):
    edge_index = knn(pos, pos, k, batch_x=batch, batch_y=batch).flip(0)#flip to get src-to-tgt
    #edge_index, _ = remove_self_loops(edge_index)#remove self loops because inv(self_dist)=too big and noisy
    src,tgt=edge_index
    diffs = pos[src]-pos[tgt]
    dists = diffs.norm(2,-1)+1e-9
    density = scatter(dists, tgt, dim=-1, reduce="sum").pow(-1)/k
    return density
```

```python
import torch
from torch_geometric.nn import MessagePassing
from torch_geometric.nn.pool import knn_graph

class NoModelChatGPT(MessagePassing):
    def __init__(self, k: int = 8, eps: float = 1e-8):
        super().__init__(aggr='sum')  # we manually normalize later
        self.k = k
        self.eps = eps

    def forward(self, K: torch.Tensor, V: torch.Tensor, batch: torch.Tensor = None) -> torch.Tensor:
        """
        K: [N, D]  positions
        V: [N, F]  vectors at those positions
        batch: [N]  batch indices (optional, for batched graphs)
        """
        # 1) build kNN graph on positions
        edge_index = knn_graph(K, k=self.k, batch=batch, loop=False)

        # 2) propagate V weighted by 1/dist
        weighted_sum = self.propagate(edge_index, x=V, pos=K)

        # 3) propagate weights alone to get normalization factor
        ones = torch.ones((V.size(0), 1), device=V.device)
        weight_sum = self.propagate(edge_index, x=ones, pos=K)

        # 4) normalize
        return weighted_sum / (weight_sum + self.eps)

    def message(self, x_j: torch.Tensor, pos_i: torch.Tensor, pos_j: torch.Tensor) -> torch.Tensor:
        """
        x_j: [E, F]      neighbor features
        pos_i: [E, D]    receiver positions
        pos_j: [E, D]    sender positions
        """
        # compute Euclidean distance
        dist = torch.norm(pos_i - pos_j, p=2, dim=-1, keepdim=True)  # [E,1]
        w = 1.0 / (dist + self.eps)  # [E,1]
        return w * x_j              # [E,F]
```

```python
from typing import Optional

import torch
from torch import Tensor
from torch.nn import Parameter

from torch_geometric.nn.conv import MessagePassing
from torch_geometric.nn.dense.linear import Linear
from torch_geometric.nn.inits import zeros
from torch_geometric.typing import (
    Adj,
    OptPairTensor,
    OptTensor,
    SparseTensor,
    torch_sparse,
)
from torch_geometric.utils import add_remaining_self_loops
from torch_geometric.utils import add_self_loops as add_self_loops_fn
from torch_geometric.utils import (
    is_torch_sparse_tensor,
    scatter,
    spmm,
    to_edge_index,
)
from torch_geometric.utils.num_nodes import maybe_num_nodes
from torch_geometric.utils.sparse import set_sparse_value
class Interpolator(MessagePassing):

    def __init__(
        self,
        in_channels: int,
        out_channels: int,
        improved: bool = False,
        cached: bool = False,
        **kwargs,
    ):
        kwargs.setdefault('aggr', 'add')
        super().__init__(**kwargs)

        if add_self_loops is None:
            add_self_loops = normalize

        if add_self_loops and not normalize:
            raise ValueError(f"'{self.__class__.__name__}' does not support "
                             f"adding self-loops to the graph when no "
                             f"on-the-fly normalization is applied")

        self.in_channels = in_channels
        self.out_channels = out_channels
        self.improved = improved
        self.add_self_loops = add_self_loops
        self.normalize = normalize

        self.lin = Linear(in_channels, out_channels, bias=False,
                          weight_initializer='glorot')


        self.reset_parameters()

    def reset_parameters(self):
        super().reset_parameters()
        self.lin.reset_parameters()
        zeros(self.bias)

    def forward(self, x: Tensor, edge_index: Adj,
                edge_weight: OptTensor = None) -> Tensor:

        x = self.lin(x)

        # propagate_type: (x: Tensor, edge_weight: OptTensor)
        out = self.propagate(edge_index, x=x, edge_weight=edge_weight)

        return out


    def message(self, x_j: Tensor, edge_weight: OptTensor) -> Tensor:
        return x_j if edge_weight is None else edge_weight.view(-1, 1) * x_j

    def message_and_aggregate(self, adj_t: Adj, x: Tensor) -> Tensor:
        return spmm(adj_t, x, reduce=self.aggr)
```
